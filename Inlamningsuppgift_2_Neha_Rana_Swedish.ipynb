### Part 1: Numpy

```python
import numpy as np
```

Konstruera en array med 10 kolumner och 5 rader av float-tal. Inget tal får vara samma.

```python
first_array= np.arange(1, 51).reshape(5,10)
first_array = first_array.astype('float64')
print(first_array)
```

Konstruera en ny array till med 10 kolumner och 7 rader av float-tal. Den skall inte vara samma som ovan array..

```python
second_array= np.arange(51, 121).reshape(7,10)
second_array= second_array.astype('float64')
print(second_array)
```

Gör om de ovan arrayerna så att de har 5 kolumner men innehåller samma data.

```python
first_array_reshaped= first_array.reshape(10,5)
first_array_reshaped
```

```python
second_array_reshaped= second_array.reshape(14,5)
second_array_reshaped
```

Slå ihop de bägge arrayerna till en array.

```python
arr= np.concatenate((first_array_reshaped, second_array_reshaped), axis= 0)
print(arr)
```

Ändra ordningen på kolumnerna så att de har ordningen kolumn5, kolumn3, kolumn4, kolumn1 och kolumn2.

```python
desired_index= [4, 2, 3, 0, 1] # As index starts from 0, so column 5 has ndex 4
desired_array= arr[:,desired_index]
print(desired_array)
```

Så bättre att gruppera dem kontinentmässigt. Så i koden nedan, gör jag grupper 'continent wise'. Jag tog 'continent_countrycode_mapping.csv' fån nätet för att mappa 'continents' till 'country codes'. Jag lägde till 'Continent kolumn' i alla 4 dataseter.

```python
import csv
continent_count_data = pd.read_csv('continent_countrycode_mapping.csv')

mydict = {}
with open('continent_countrycode_mapping.csv', newline='') as f:
    reader = csv.reader(f)
    mydict = {rows[0]: rows[1] for rows in reader}

def GetContinent(counryCode):
    try:
        return mydict[counryCode]
    except KeyError:
        return None

data1['Continent'] = data1['Code'].apply(lambda x: GetContinent(x))
```
